```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sp5 = yf.Ticker("^GSPC")
hist_sp5 = sp5.history(start = "2015-09-01", end = "2020-09-30", interval = "1mo", auto_adjust = True, actions = False)
hist_sp5.describe()
```

```python
hist_sp5.dropna(axis=0, subset=["Close"], inplace=True)
hist_sp5.describe()
```

```python
hist_sp5.head(15)
```

```python
nflx = yf.Ticker("NFLX")
hist_nf = nflx.history(start = "2015-09-01", end = "2020-09-30", interval = "1mo", auto_adjust = True, actions = False)
hist_nf.describe()
```

```python
hist_nf.dropna(axis=0, subset=["Close"], inplace=True)
hist_nf.tail()
```

```python
hist_nf.head(15)
```

```python
#amazon
amz = yf.Ticker("AMZN")
hist_am = amz.history(start = "2015-09-01", end = "2020-09-30", interval = "1mo", auto_adjust = True, actions = False)
hist_am.dropna(axis=0, subset=["Close"], inplace=True)
hist_am.tail()
```

```python
#building a table with the closing prices only
prices = pd.concat([hist_sp5['Close'], hist_nf['Close'], hist_am['Close']], axis=1)
prices.columns = ['S&P500', 'Netflix', 'Amazon']
prices.describe()
```

```python
prices.tail()
```

```python
#a quick look in prices before getting the variation
prices.plot(
    style=['blue', 'red', 'gold']
)
```

```python
#monthly returns
mon_returns = prices.pct_change()
mon_returns.dropna(axis=0, inplace= True)
mon_returns.describe()
```

```python
#graphically
mon_returns.plot(
    title = 'S&P500, Netflix and Amazon Monthly Returns - 5 years monthly',
    figsize = [15,8],
    style = ['royalblue', 'salmon', 'yellow'],
    grid = True,
    subplots = True,
    sharey = True,
    sharex = True)
```

```python
def beta(stock_monthly_returns):
    x = mon_returns['S&P500']
    y = stock_monthly_returns
    model = stats.linregress(x, y)
    return print("beta =", "%.2f" % model[0])
```

```python
beta(mon_returns['Netflix'])
```

```python
beta(mon_returns['Amazon'])
```

Got really close to Netflix beta as shown in yahoo finance.
