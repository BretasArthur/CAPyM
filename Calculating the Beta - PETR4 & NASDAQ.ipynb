```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
nasdaq = yf.Ticker("^IXIC")
hist_nasdaq = nasdaq.history(start="2015-10-01", end = "2020-10-01", interval = "1mo")
hist_nasdaq.describe()
```

```python
hist_nasdaq.dropna(axis=0, subset=["Close"], inplace=True)
hist_nasdaq.describe()
```

```python
hist_nasdaq.head(15)
```

```python
petr4 = yf.Ticker("PETR4.SA")
hist_p4 = petr4.history(start="2015-10-01", end = "2020-10-01", interval = "1mo")
hist_p4.describe()
```

```python
hist_p4.dropna(axis=0, subset=['Close'], inplace=True)
hist_p4.describe()
```

```python
hist_p4.head(15)
```

```python
prices = pd.concat([hist_nasdaq['Close'], hist_p4['Close']], axis=1)
prices.columns = ['NASDAQ', 'Petrobras']
prices.describe()
```

```python
prices.head()
```

```python
prices.plot(
    style=['red', 'green']
)
```

```python
mon_returns = prices.pct_change()
mon_returns.dropna(axis=0, inplace= True)
mon_returns.head()
```

```python
mon_returns.plot(
    title = 'Petrobrs and NASDAQ Montly Returns - from november 2015 to octuber 2020',
    figsize = [15,8],
    style = ['salmon', 'limegreen'],
    grid = True)
```

```python
x = mon_returns['NASDAQ']
y = mon_returns['Petrobras']

model = stats.linregress(x, y)

model
```

```python
print("Beta =", model[0])
```
