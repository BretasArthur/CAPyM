```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
ibov = yf.Ticker("BOVA11.SA")
hist_b11 = ibov.history(start="2015-11-01", end = "2020-10-01", interval = "1mo")
hist_b11.describe()
```

```python
hist_b11.dropna(axis=0, subset=["Close"], inplace=True)
hist_b11.describe()
```

```python
hist_b11.head(15)
```

```python
petr4 = yf.Ticker("PETR4.SA")
hist_p4 = petr4.history(start="2015-11-01", end = "2020-10-01", interval = "1mo")
hist_p4.describe()
```

```python
hist_p4.dropna(axis=0, subset=['Close'], inplace=True)
hist_p4.describe()
```

```python
hist_p4.head(15)
```

```python
prices = pd.concat([hist_b11['Close'], hist_p4['Close']], axis=1)
prices.columns = ['Petrobrás', 'iBovespa']
prices.describe()
```
