```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sp5 = yf.Ticker("^GSPC")
hist_sp5 = sp5.history(start = "2015-09-01", end = "2020-09-30", interval = "1mo", auto_adjust = True, actions = False)
hist_sp5.describe()
```

```python
hist_sp5.dropna(axis=0, subset=["Close"], inplace=True)
hist_sp5.describe()
```

```python
hist_sp5.head(15)
```

```python
nflx = yf.Ticker("NFLX")
hist_nf = nflx.history(start = "2015-09-01", end = "2020-09-30", interval = "1mo", auto_adjust = True, actions = False)
hist_nf.describe()
```

```python
hist_nf.dropna(axis=0, subset=["Close"], inplace=True)
hist_nf.tail()
```

```python
hist_nf.head(15)
```

```python
prices = pd.concat([hist_sp5['Close'], hist_nf['Close']], axis=1)
prices.columns = ['S&P500', 'Netflix']
prices.describe()
```

```python
prices.tail()
```

```python
prices.plot(
    style=['blue', 'red']
)
```

```python
mon_returns = prices.pct_change()
mon_returns.dropna(axis=0, inplace= True)
mon_returns.head()
```

```python
mon_returns.plot(
    title = 'Netflix and S&P500 Montly Returns - 5 years monthly',
    figsize = [15,8],
    style = ['royalblue', 'salmon'],
    grid = True)
```

```python
x = mon_returns['S&P500']
y = mon_returns['Netflix']

model = stats.linregress(x, y)

model
```

```python
print("beta =", "%.2f" % model[0])
```

Got really close to Netflix beta as shown in yahoo finance.
