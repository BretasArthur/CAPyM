```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
#from scipy import stats
import statsmodels.api as sm
```

```python
sp5 = yf.Ticker("^GSPC")
hist_sp5 = sp5.history(period='5y', interval = "1mo")
hist_sp5.describe()
```

```python
hist_sp5.dropna(axis=0, subset=["Close"], inplace=True)
hist_sp5.describe()
```

```python
hist_sp5.head(15)
```

```python
petr4 = yf.Ticker("PETR4.SA")
hist_p4 = petr4.history(period='5y', interval = "1mo")
hist_p4.describe()
```

```python
hist_p4.dropna(axis=0, subset=['Close'], inplace=True)
hist_p4.describe()
```

```python
hist_p4.head(15)
```

```python
prices = pd.concat([hist_sp5['Close'], hist_p4['Close']], axis=1)
prices.columns = ['S&P500', 'Petrobras']
prices.describe()
```

```python
prices.tail()
```

```python
prices.plot(
    style=['gold', 'green']
)
```

```python
mon_returns = prices.pct_change()
mon_returns.dropna(axis=0, inplace= True)
mon_returns.head()
```

```python
mon_returns.plot(
    title = 'Petrobrs and S&P500 Montly Returns - from november 2015 to octuber 2020',
    figsize = [15,8],
    style = ['yellow', 'limegreen'],
    grid = True)
```
